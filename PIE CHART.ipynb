```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('employee.csv')
data.shape
```

```python
plt.pie(data['BusinessTravel'].value_counts(), labels=data['BusinessTravel'].value_counts().index)
```

```python
data['BusinessTravel'].value_counts().index
```

```python
plt.pie(data['BusinessTravel'].value_counts(), labels=data['BusinessTravel'].value_counts().index)
my_circle=plt.Circle((0,0),0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

slices = [60,40, 30, 20]
labels = ['Sixty', 'Forty', 'Thirty', 'Twenty']
colors = ['#008fd5', '#fc4f30', '#e5ae37', '#6d904f']

plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
plt.style.use("fivethirtyeight")

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0.2, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, explode=explode, shadow=True,
        startangle=90, autopct='%1.2f%%',
        wedgeprops={'edgecolor': 'black'})

plt.title("Stackoverflow Developer Survey")
plt.tight_layout()
plt.show()
```

```python
plt.style.use("fivethirtyeight")

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, explode=explode, shadow=True,
        startangle=90, autopct='%1.1f%%',
        wedgeprops={'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```
