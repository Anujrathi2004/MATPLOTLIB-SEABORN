```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style='dark')
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, height=4, aspect=3)
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x='total_bill', y='tip',hue='sex', data=tips, height=4, aspect=3)
```

```python
tips['size'].value_counts(ascending=True)
```

```python
sns.relplot(x='total_bill', y='tip',hue='size', data=tips, height=6, aspect=2)
```

```python
sns.relplot(x='total_bill', y='tip',style='sex',hue='size', data=tips, height=4, aspect=3)
```

```python
sns.relplot(x='total_bill', y='tip',style='time',size='size',hue='smoker', data=tips, height=5, aspect=2)
```

```python
sns.relplot(x='total_bill', y='tip',col='sex',hue='smoker', data=tips, height=5, aspect=1)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
plt.figure(figsize=(20,6))
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='smoker', size='size', style='time')
```

```python
sns.countplot(x=tips['sex'])
```

```python
sns.catplot(x='day', y= 'total_bill', data=tips, height=5, aspect=2.5)
```

```python
sns.catplot(x='day', y= 'tip', data=tips, height=5, aspect=2.5, hue='sex')
```

```python
diamonds=sns.load_dataset('diamonds')
diamonds.head()
```

```python
sns.catplot(x='day', y= 'total_bill',hue='sex', kind='violin', data=tips, height=5, aspect=2.5)
```
