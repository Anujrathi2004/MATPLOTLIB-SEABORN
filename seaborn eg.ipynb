```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
loan=pd.read_csv('train1.csv')
```

```python
loan.head()
```

```python
plt.figure(figsize=(15,6))
sns.countplot(x='Gender', data=loan)
```

```python
plt.figure(figsize=(15,6))
sns.countplot(x='Gender', data=loan, hue='Loan_Status', palette='husl')
```

```python
plt.figure(figsize=(15,6))
sns.countplot(x='Married', hue='Loan_Status', data=loan)
```

```python
plt.figure(figsize=(15,6))
sns.boxplot(x='Loan_Status', y='ApplicantIncome', data=loan)
```

```python
plt.figure(figsize=(15,6))
sns.boxplot(x='Loan_Status', y='CoapplicantIncome', data=loan)
```

```python
plt.figure(figsize=(15,6))
sns.countplot(x='Property_Area', hue='Loan_Status', data=loan)
```
